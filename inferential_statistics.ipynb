```python
import numpy as np
import scipy.stats as stats
import math
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

## Pruebas de hipotesis

$$ (\overline{x}_1 - \overline{x}_2)\pm t_{df} \times SE_{(\overline{x}_1 - \overline{x}_2)}

$$ SE_(\overline{x}_1 - \overline{x}_2) = \sqrt{\frac{S^2_1}{n_1} + \frac{S^2_2}{n_2}}

$$ dF = min(n_1 -1, n_2 - 1)

```python
media_universal=25
media_H0=22
sd=6
n=64

Z=(media_universal-media_H0)/(sd/math.sqrt(n))
print("Estadistico del set", Z)

alpha=0.05
print(alpha)
```

```python
1-stats.norm.cdf(Z)
```

```python
alpha=1-0.05
z_critico=stats.norm.ppf(alpha)
z_critico
```

```python
critical_value=z_critico

samples=100
x_plot=np.linspace(-5,5,samples)
y_plot=stats.norm.pdf(x_plot,0,1)
plt.plot(x_plot,y_plot)

x_crit=np.linspace(critical_value,5,samples)
y_crit=stats.norm.pdf(x_crit,0,1)
plt.fill_between(x_crit,y_crit,alpha=.5)

plt.plot([Z,Z],[0,stats.norm.pdf(Z)])

plt.legend(['Z critico', 'Z score'])
```
